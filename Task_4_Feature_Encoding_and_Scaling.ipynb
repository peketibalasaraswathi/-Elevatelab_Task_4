```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

```

```python
df = pd.read_csv("adult.csv")
df.head()
```

```python
df.shape
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

print("Categorical:", categorical_features)
print("Numerical:", numerical_features)
```

```python
df = df.replace(' ?', np.nan)
df.dropna(inplace=True)
```

```python
le = LabelEncoder()
df['income'] = le.fit_transform(df['income'])
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_features.drop('income'))
```

```python
df_encoded.head()
```

```python
scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(df_encoded[numerical_features])
```

```python
df[numerical_features].describe()
```

Feature scaling ensures that all numerical features contribute equally to the model.
Algorithms like KNN, SVM, Logistic Regression, and Gradient Descent perform poorly when features have different scales.
StandardScaler transforms data to zero mean and unit variance, improving convergence speed and accuracy.

```python
df_encoded.to_csv("adult_processed.csv", index=False)
```
